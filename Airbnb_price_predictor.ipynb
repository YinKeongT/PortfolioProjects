```python

import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
# read file airbnb.csv to train_data
train_data = pd.read_csv("airbnb.csv")
```

```python
# drop values that are NaN
train_data = train_data.drop(["300"],axis=1)
```

```python
train_data
```

```python
"""
How to remove NaN values
"""
def clean(data):    
    cols = data.head()
    for col in cols:
        data[col].fillna(data[col].median(), inplace=True)
    return data

data = clean(train_data)
```

```python
data
```

```python
target = data["250"]
```

```python
target
```

```python
# create train_data by mapping features that want to be used for training model and dropping target feature
# price when available unit = 250

train_data = train_data.drop(["250"],axis=1)
```

```python
# check whether imported features are correct or not
train_data
```

```python
# Set target variable to price when available unit = 250

target = data["250"]
```

```python
# check whether target variable is set correctly or not
target
```

```python
# Train model using Random Forest Classifier by setting Number of Trees
# n_estimators=100, Depth of Trees max_depth=5, and Set random_state=1 to prevent randomness in result

model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
```

```python
# Fit model on training data X that contains the features we have chosen and target (price when there are 250 available units)
# using model.fit()

model.fit(train_data,target)
```

```python
# Predict price when there are 250 available units on test data using the trained model using model.predict
test_data = pd.read_csv("test_airbnb.csv")
print(test_data)
predictions = model.predict(test_data)
predictions
```

```python
# Create output in form of table using pd.Dataframe

output = pd.DataFrame(
    {'day to check-in': test_data.day_to_checkin, 
     'price': predictions}
)
output
```

```python
# save output file to test1.csv by using output.to_csv and setting no index by index=False 

output.to_csv('test1.csv', index=False)
```
